botsdf has the field iso_code. This is just a code for countries. It does not contribute much to the data, so it is dropped along with continent and location since we already know all the data is for Botswana.

```python
botsdf['iso_code'].nunique()
```

```python
bots2df = botsdf.drop('iso_code', axis=1)
```

```python
bots2df = botsdf.drop(['iso_code','continent', 'location'], axis=1)
```
